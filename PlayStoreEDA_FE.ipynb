<h3>Replacing the Mb with 000 and kb with empty string to make the Size column as integer</h3>

Replacing 'Varies with device' With Nan Value 

```python
df_copy['Size']=df_copy['Size'].str.replace('Varies with device',str(np.nan))
```

Check the distribution of the numeric variable

```python
num_df=df[numeric_features]
cat_df=df[categorical_features]
```

```python
num_df.head()
```

```python
sns.kdeplot(num_df['Rating'])
```

```python
for feature in numeric_features:
  sns.kdeplot(df[feature])
  plt.Figure(figsize=(10,10))
  plt.xlabel(feature)
  plt.ylabel("Count")
  plt.title("Numerical Feature")
  plt.show()
```

```python
plt.figure(figsize=(15,15))
plt.suptitle("Univariate Analysis of Numerical Features",fontsize=20,fontweight='bold',alpha=0.8,y=1.)

for i in range(0,len(numeric_features)):
  plt.subplot(5,3,i+1)
  sns.kdeplot(x=df[numeric_features[i]],shade=True,color='r')
  plt.xlabel(numeric_features[i])
  plt.tight_layout()
```
