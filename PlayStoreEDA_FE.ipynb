<h3>Replacing the Mb with 000 and kb with empty string to make the Size column as integer</h3>

```python
df_copy['Size']=df_copy['Size'].str.replace('M','000')
df_copy['Size']=df_copy['Size'].str.replace('k','')
```

```python
df_copy['Size'].head()
```

```python
df_copy['Size'].unique()
```

Replacing 'Varies with device' With Nan Value 

```python
df_copy['Size']=df_copy['Size'].str.replace('Varies with device',str(np.nan))
```

```python
df_copy['Size'].unique()
```

<h3>Converting the Size column to float type</h3>

```python
df_copy['Size']=df_copy['Size'].astype('float')
```
